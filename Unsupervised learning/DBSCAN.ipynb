```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Mall_Customers.csv")
df
```

```python
features=["Annual Income (k$)","Spending Score (1-100)"]
x=df[features]
```

```python
x_standardized=StandardScaler().fit_transform(x)
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
clusters=dbscan.fit_predict(x_standardized)
```

```python
centroids=pd.DataFrame(x_standardized).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=clusters,cmap='viridis',edgecolors='pink',s=50)
plt.title('DBSCAN Clustering of Mall customer date')
plt.xlabel("Annual   Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```
