```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("/content/Iris.csv")
df
```

```python
features=["SepalLengthCm","SepalWidthCm"]
x=df[features]
```

```python
x_standardized=StandardScaler().fit_transform(x)
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
clusters=dbscan.fit_predict(x_standardized)
```

```python
centroids=pd.DataFrame(x_standardized).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(7,6))
plt.scatter(x['SepalLengthCm'],x['SepalWidthCm'],c=clusters,cmap='viridis',edgecolors='pink',s=50)
plt.title('DBSCAN IRIS SPECIES')
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.show()
```
