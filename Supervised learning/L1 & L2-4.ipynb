```python
#ridge regression
#    L2 regularization
#    add penalty term to error term
#    penalty term is in square
#    ridge(B)=∑(i=1)(yi-xi^2B)^2+lambda∑(j=1)WjBj^2
#    does reach zero but can reach zero(coz square term wont reach zero)
#    diamond
```

```python
#LASS0 REGRESSION
#    L1 regularization
#    add penalty term to error term
#    penalty term (LS obj=alp*(sum of coeffiecients))[absolute values as coeff]
#    Llasso(B)=∑i=1(yi-xi^2B)^2+lambda∑(j=1)|Bj|
#    for absolute errror(0) since it has modulus(modulus can reach zero)
#    circle
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2.2,3.2,4.2,5.6,11.2])
```

```python
x_b=np.c_[np.ones((x.shape[0],1)),x]
x_b
```

```python
alpha_ridge=1.0
alpha_lasso=0.1
num_iterations=1000 #num_iterations: This is the number of iterations or epochs
# that will be used in the optimization process.

learning_rate=0.01 #: This is a hyperparameter that determines the step size at
#which the model's parameters are updated during each iteration of the optimization process.
```

```python
def lasso_regression(x,y,alpha,num_iterations,learning_rate):
  m,n=x.shape
  theta=np.zeros(n)
  for _ in range(num_iterations):
    y_pred=x.dot(theta)
    gradient=(x.T.dot(y_pred -y)+alpha*np.sign(theta))/m #+penalty value
    theta-=learning_rate*gradient
  return theta
```

```python
def ridge_regression(x,y,alpha,num_iterations,learning_rate):
  m,n=x.shape
  theta=np.zeros(n) #keeping penalty term as 0
  for _ in range(num_iterations):
    y_pred=x.dot(theta)
    gradient=(x.T.dot(y_pred -y)+alpha*theta)/m
    theta-=learning_rate*gradient #This updates the parameter vector theta
# in the direction that minimizes the loss.
  return theta
```

```python
theta_ridge=ridge_regression(x_b,y,alpha_ridge,num_iterations,learning_rate)
theta_lasso=lasso_regression(x_b,y,alpha_lasso,num_iterations,learning_rate)
```

```python
plt.scatter(x,y,label="Data")
plt.plot(x,x_b.dot(theta_ridge),label=f'Ridge (L2) Regression',color='red')
plt.plot(x,x_b.dot(theta_lasso),label=f'Lasso (L1) Regression',color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
