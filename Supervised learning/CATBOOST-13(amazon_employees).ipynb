```python
y_train=train["ACTION"]
```

```python
x_train=train[["RESOURCE","MGR_ID","ROLE_ROLLUP_1","ROLE_ROLLUP_2","ROLE_DEPTNAME","ROLE_TITLE","ROLE_FAMILY_DESC","ROLE_CODE"]]
```

```python
cat_features=list(range(0,x_train.shape[1]))
print(cat_features)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, random_state=0,test_size=0.3)
```

```python
from catboost import CatBoostClassifier
```

```python
learning_rates=[0.01,0.1,0.25,0.5,0.75,1]
for learning_rate in learning_rates:
    cb = CatBoostClassifier(n_estimators=5, learning_rate = learning_rate, max_depth = 2, random_state = 0,cat_features=cat_features)
    cb.fit(x_train, y_train)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(cb.score(x_train, y_train)))
    print("Accuracy score (validation): {0:.3f}".format(cb.score(x_test, y_test)))

```
