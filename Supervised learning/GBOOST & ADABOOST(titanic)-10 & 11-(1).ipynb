```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
test=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
train=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
```

```python
train.set_index("PassengerId",inplace=True)#set_index() function in pandas is used to set one or more columns of a DataFrame as the index
test.set_index("PassengerId",inplace=True)
```

```python
y_train=train["Survived"]
```

```python
train.drop(labels="Survived",axis=1,inplace=True)
```

```python
train.shape
```

```python
train_test=train.append(test) #increasing the dataset since we usually take large datasets for these algorithms
train_test
```

```python
train_test.shape
```

```python
train_test.isnull().sum()
```

```python
## delete columns that are not used as features for training and prediction
columns_to_drop=["Name","Age","SibSp","Ticket","Cabin","Parch","Embarked"]
train_test.drop(labels=columns_to_drop,axis=1,inplace=True)
```

```python
train_test
```

```python
#one hot encoding
train_test_dummies=pd.get_dummies(train_test,columns=["Sex"])
```

```python
train_test_dummies
```

```python
train_test_dummies.isnull().sum()
```

```python
train_test_dummies.fillna(value=0.0,inplace=True)
```

```python
x_train=train_test_dummies.values[0:891] #taken as train dt
x_test=train_test_dummies.values[891:] #taken as test dt
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train_scale=scaler.fit_transform(x_train) #MinMaxScaler,the fit_transform() method is used on the training data, which calculates the minimum and maximum values of each feature in the training set and then scales the training data accordingly.
X_test_scale=scaler.transform(x_test)#scaler.transform() is used on the test data to scale it based on the same minimum and maximum values calculated from the training data using fit_transform().
```

```python
# split training feature and target sets into training and validation subsets
from sklearn.model_selection import train_test_split
x_train_sub, x_validation_sub, y_train_sub, y_validation_sub = train_test_split(X_train_scale, y_train, random_state=0)
```

```python

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
learning_rates=[0.01,0.1,0.25,0.5,0.75,1]
for learning_rate in learning_rates:
    gb = GradientBoostingClassifier(n_estimators=20, learning_rate = learning_rate, max_features=2, max_depth = 2, random_state = 0)
    gb.fit(x_train_sub, y_train_sub)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb.score(x_train_sub, y_train_sub)))
    print("Accuracy score (validation): {0:.3f}".format(gb.score(x_validation_sub, y_validation_sub)))
```

```python
#ADA BOOST Classifier
from sklearn.ensemble import AdaBoostClassifier  # For regression tasks
from sklearn.metrics import classification_report  # For classification tasks
```

```python
ab=AdaBoostClassifier(n_estimators=500,random_state=0)
ab.fit(x_train_sub, y_train_sub)
print("Accuracy score (training): {0:.3f}".format(ab.score(x_train_sub, y_train_sub)))
print("Accuracy score (validation): {0:.3f}".format(ab.score(x_validation_sub, y_validation_sub)))
```

```python
learning_rates=[0.05,0.001,0.25,0.5,0.75,1]
for learning_rate in learning_rates:
    ab= AdaBoostClassifier(n_estimators=10000,random_state = 0)
    ab.fit(x_train_sub, y_train_sub)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(ab.score(x_train_sub, y_train_sub)))
    print("Accuracy score (validation): {0:.3f}".format(ab.score(x_validation_sub, y_validation_sub)))
```
