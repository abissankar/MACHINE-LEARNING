```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/1.01. Simple linear regression.csv") #read the csv file
data
```

```python
x_data=data.iloc[:10,0:1].values #specifying x data
```

```python
y_data=data.iloc[:10,1:2].values #specifying y data
```

```python
from sklearn.linear_model import LinearRegression #for linear regression importing from sklearn
model = LinearRegression() #calling linearreg package
```

```python
model.fit(x_data,y_data)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_data,y_data,color="red")  #scatter plot to know where the data points are actually located
plt.plot(x_data,model.predict(x_data),color='blue') #plotting the linear line
plt.title("Linear regression")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
