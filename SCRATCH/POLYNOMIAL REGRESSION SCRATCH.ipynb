```python
#Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data.
#linear regression works only for linear data either increasing or decreasing
#for non linear data polynomial regresiion(in this it is converted into linear using some methods like multiplying wit degree)
#y=bo+bx1(L)  y=bo+bx1+bx1^2(P)
#quad=deg 2
#cub=deg 3
#hyperparameter-power(additional parameter for tuning data)(given by us)
#parameter-training data(pre)
#degree selection-1]forward selection-increase the deg parameter to optimal results
                # 2]backward selection-decrease
#polynomial linear reg- because here the poly is conv to linear
#simple lin reg-deg of 1(y=bo+bx1)
#mul lin reg-deg of 1(y=bo+bx1+bx2...bxn)
#poly lin reg-deg of n(y=bo+bx1+bx1^2...bx^n)
#overfit- training optimal testing not optimal(the model gets used)#mugging
#underfit-training also not work and testing also not optimal
#loss = single (y-y^)^2
#cost = sum of all (E(y-y^)^2/n)
#variance -[using complex feature calculated ](when the data is complex the model learns too much from the model)(takes the noise also into  account)
#bias=actual-predicted(error)[using simple feature calculated]{these assump may not always be correct}
#bias variance trade off -to get optimal model#(low variance low bias-perfect)
                                              #(low variance high bias-overfit)
                                              # (high variance high bias-underfit)
#nd poly-y=a0+a1x1+a2x1^2....+anxn
#2 deg poly-y=a0+a1x1+a2x1^2
#3 deg poly-y=a0+a1x1+
#data->poly tranf->lin reg-> prediction
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#y=a0-a1x1-a2x1^2
x=np.array([1,2,3,4,5])
y=np.array([2.1,3.8,4.5,5.2,9.3])
```

```python
degree =2
```

```python
x_poly=np.column_stack([x**i for i in range(degree+1)])# to create 2d array from 1d
x_poly #x^0 x^1 x^2
```

```python
#weights assignment(gd - changes m&c)
weights=np.zeros(degree+1)
weights #a0,a1,a2
```

```python
#assign alpha(learning rate and epoch)
learning_rate=0.001 #default
num_iterations= 10000 #epoch

```

```python
for _ in range(num_iterations):
  #calculate predic
  predictions=np.dot(x_poly,weights)
  #calc the error
  error=predictions-y
  #update the weights (coeff ) using gradient descent
  gradient=np.dot(x_poly.T,error)/len(x) # T-transpose mx====y-y^/n
  weights-=learning_rate*gradient #The weights are adjusted in the opposite direction of the gradient in order to minimize the loss.
  print(f'weights for iterations{_}:{weights}')
```

```python
#predictions
def predict(x,weights):
  x_poly=np.array([x**i for i in range(degree+1)])
  return np.dot(x_poly,weights)
```

```python
plt.scatter(x,y,label="Data points")
x_range=np.linspace(min(x),max(x),100)
y_pred =[predict(x,weights) for x in x_range]
plt.plot(x_range,y_pred,color='r',label="polynomial regression")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
